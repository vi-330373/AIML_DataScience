## â€¢	Missing value treatment: Check which variables have missing values and use appropriate treatments. 
### For each of the variables, find the number of missing values and provide the value that they have been imputed with.


```python
##Find the Number of missing values in each column:
df1.isna().sum()
```

```python
df2=df1.dropna(axis=0, how='any')
df2.shape
## If we directly remove all the null values from the sheet, then we will be left with only 8887 rows of data.
```

```python
##Substituting the missing value in Column 'Price' with the median:
df1['Price'].fillna(value=df1['Price'].mean(),inplace=True)
```

```python
df1.isnull().sum()
```

```python
df_2=df1.fillna(value='?')
```

```python
pd.crosstab(df_2['Rooms'],df_2['Bedroom2'])

#As per the heatmap, Rooms and BedRooms column are highly related. Same thing can be inferred from the below cross tab. Hence replace the missing values in Bedroom and Bathroom column with no of Rooms.
```

```python
pd.crosstab(df_2['Bedroom2'],df_2['Bathroom'])
```

```python
df_2.Bedroom2.replace('?', df_2.Rooms, inplace=True)
df_2.Bathroom.replace('?', df_2.Bedroom2, inplace=True)
```

```python
df1['Bedroom2']=df_2['Bedroom2']
df1['Bathroom']=df_2['Bathroom']
```

```python
##Substituting the missing value in other Columns with the median:
df1['Car'].fillna(value=df1['Car'].median(),inplace=True)
df1['Landsize'].fillna(value=df1['Landsize'].median(),inplace=True)
df1['BuildingArea'].fillna(value=df1['BuildingArea'].median(),inplace=True)
df1['YearBuilt'].fillna(value=df1['YearBuilt'].median(),inplace=True)
df1['Lattitude'].fillna(value=df1['Lattitude'].median(),inplace=True)
df1['Longtitude'].fillna(value=df1['Longtitude'].median(),inplace=True)
```

```python
## dropping of the rows with left out missing values:
df1.dropna(inplace=True)
```

â€¢	Outlier Treatment: 
		Identify the variables : Make a subset of the dataset with all the numeric variables. 
		Outliers : For each variable of this subset, carry out the outlier detection. Find out the percentile distribution of each variable and carry out capping and flooring for outlier values.  

```python
Q1 = df1.quantile(0.25)
Q3 = df1.quantile(0.75)
IQR = Q3 - Q1 ##InterQuartile Range for all the columns
print(IQR)
```

```python
sns.pairplot(df1)
```

```python
##Outliers treatment in Price:
sns.boxplot(x=df1['Price'])
```

```python
Q1 = df1['Price'].quantile(0.25)
Q3 = df1['Price'].quantile(0.75)
IQR = Q3 - Q1
outliers_price=(df1['Price'] < (Q1 - 1.5 * IQR))|(df1['Price'] > (Q3 + 1.5 * IQR))
```

```python
outliers_price.value_counts()
```

```python
Outlier_df=df1[outliers_price==True]##removing the outliers in price
```

```python
Outlier_df.shape
```

```python
##Outliers treatment in landsize:
sns.boxplot(x=df1['Landsize'])
Q1 = df1['Landsize'].quantile(0.25)
Q3 = df1['Landsize'].quantile(0.75)
IQR = Q3 - Q1
outliers_Landsize=(df1['Landsize'] < (Q1 - 1.5 * IQR))|(df1['Landsize'] > (Q3 + 1.5 * IQR))
print(outliers_Landsize.value_counts())
Outlier_df=Outlier_df.append(df1[outliers_Landsize==True])##removing the outliers in Landsize
print('shape of outliers after adding outliers from Landsize:',Outlier_df.shape)
```

```python
##Outliers treatment in Rooms:
sns.boxplot(x=df1['Rooms'])
Q1 = df1['Rooms'].quantile(0.25)
Q3 = df1['Rooms'].quantile(0.75)
IQR = Q3 - Q1
outliers_Rooms=(df1['Rooms'] < (Q1 - 1.5 * IQR))|(df1['Rooms'] > (Q3 + 1.5 * IQR))
print(outliers_Rooms.value_counts())
##Not removing these outliers
```

```python
Outlier_df.shape
```

```python
##Outliers treatment in Bedroom2:
sns.boxplot(x=df1['Bedroom2'])
Q1 = df1['Bedroom2'].quantile(0.25)
Q3 = df1['Bedroom2'].quantile(0.75)
IQR = Q3 - Q1
outliers_Bedroom2=(df1['Bedroom2'] < (Q1 - 1.5 * IQR))|(df1['Bedroom2'] > (Q3 + 1.5 * IQR))
print(outliers_Bedroom2.value_counts())
#Outlier_df=Outlier_df.append(df1[outliers_Bedroom2==True])##adding the outliers from Bedrooms
#print('shape of outliers after adding outliers from Bedroom2:',Outlier_df.shape)

##Not removing these outliers
```

```python
##Outliers treatment in Propertycount:
sns.boxplot(x=df1['Propertycount'])
Q1 = df1['Propertycount'].quantile(0.25)
Q3 = df1['Propertycount'].quantile(0.75)
IQR = Q3 - Q1
outliers_Propertycount=(df1['Propertycount'] < (Q1 - 1.5 * IQR))|(df1['Propertycount'] > (Q3 + 1.5 * IQR))
print(outliers_Propertycount.value_counts())
Outlier_df=Outlier_df.append(df1[outliers_Propertycount==True])##adding the outliers from Propertycount
print('shape of outliers after adding outliers from Propertycount:',Outlier_df.shape)
```

```python
#Final df after removing the outliers:
df1=df1[df1.apply(lambda x: x.values.tolist() not in Outlier_df.values.tolist(), axis=1)]
```
